# Titanic Dataset

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.read_csv('titanic.csv')
```

```python
df.head()
```

```python
df.info()
```

```python
df.describe()
```

```python
df.drop('Cabin', axis=1, inplace=True)
```

```python
df.info()
```

```python
df.Age.mean()
```

```python
df.Age.fillna(df.Age.mean(), inplace=True)

df.info()
```

```python
df.columns
```

```python
df.dropna(subset=['Embarked'], inplace=True)

df.info()
```

# Feature Engineering

```python
#Deriving a field that doesn't exist in the dataset, butcan be derived from
#existing fields

df['Family_size'] = df['SibSp'] + df['Parch'] + 1
df['Name_length'] = df['Name'].apply(len)
df.head()
```

```python
# Filter over 30
df[df.Age > 30]

# Filter survived and over 30
df[(df.Survived == 1) & (df.Age > 30)]

# Filter not in 3rd Class
df[df.Pclass != 3]

# Filter not in 1st class or 3rd class
df[(df.Pclass == 1) | (df.Pclass == 3)]
```

```python
sns.kdeplot(data=df1, x="Pclass", hue="Survived", fill=True)
plt.xticks(df1['Pclass'].unique().astype(int))  # Set x-axis ticks to unique integer values of Pclass
plt.show()
```

```python
sns.barplot(x='Pclass', y='Survived', data=df1) # Changed 'PClass' to 'Pclass'
```

```python
grouped_fare = df.groupby(['Pclass'])['Fare'].mean() # Changed 'PClass' to 'Pclass'
sns.barplot(x=grouped_fare.index, y=grouped_fare.values) # Added x and y to specify data for the plot
```
